Распределение похоже на экспоненциальное.
Расчитаем lamda и size для генерации случайных чисел

```python
lambda_dist = 1 / df_dist['distance_km'].mean()
print(f"λ = {lambda_dist}")
```

```python
size_dist = len(df[df['distance_km'] <= 0])
print(f"Количество аномальных значений distance_km = {size_dist}")
```

```python
value_dist = np.random.exponential(scale=lambda_dist, size=size_dist)
```

```python
value_dist
```

```python
value_dist.max()
```

```python
value_dist.min()
```

```python
# value_dist_final = df_dist.min() + (value_dist - value_dist.min()) / (value_dist.max() - value_dist.min()) * 100
value_dist_final = 0.005 + (value_dist - 2.6563837618403634e-06) / (0.3904698612177088 - 2.6563837618403634e-06) * 100
```

```python
# Посмотрим на распределение случаных величин скорости
plt.figure(figsize=(12, 6))
plt.hist(value_dist_final, bins=1000, color='skyblue', alpha=1)

plt.xlabel('Расстояния')
# plt.ylabel('Значения')
plt.title('Распределение расстояний')
plt.legend()
plt.grid(axis='y', alpha=0.75)

plt.show()
```

```python
# Присвоим аномальным значениям distance_km значения из экспоненциального распределения
df.loc[df['distance_km'] <= 0, 'distance_km'] = value_dist_final
```

```python
df_min_mean = np.mean(df_min['duration_min'])
df_min_var = np.var(df_min['duration_min'])
```

```python
# Оценка параметра shape (к) и scale (θ) через метод моментов
shape = df_min_mean**2 / df_min_var  # shape
scale = df_min_var / df_min_mean     # scale
```

```python
# value_min = np.random.exponential(scale=lambda_1, size=size_min)
value_min = np.random.gamma(shape=shape, scale=scale, size=size_min)
```

```python
# Посмотрим на распределение случаных величин повремени
plt.figure(figsize=(12, 6))
plt.hist(value_min, bins=1000, color='skyblue', alpha=1)

plt.xlabel('Время в пути')
# plt.ylabel('Значения')
plt.title('Распределение времени')
plt.legend()
plt.grid(axis='y', alpha=0.75)

plt.show()
```

```python
import xgboost as xgb
```

```python
# Преобразуем вектор целевой переменной в формат Dmatrix, требуемый XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
# Задаем параметры модели XGBoost
params = {
    'objective': 'binary:logistic',  # Задаем целевую функцию для бинарной классификации
    'max_depth': 30,  # Максимальная глубина деревьев
    'eta': 0.01,  # Шаг градиентного спуска
    'subsample': 0.7,  # Доля объектов, используемых для обучения каждого дерева
    'colsample_bytree': 0.7,  # Доля признаков, используемых для обучения каждого дерева
    'seed': 0  # Значение сида для воспроизводимости результатов
}
```

```python
# Обучаем модель XGBoost на обучающей выборке
model = xgb.train(params, dtrain, num_boost_round=100)
```

```python
# Предсказываем значения целевой переменной на тестовой выборке
y_pred = model.predict(dtest)

# Преобразуем вероятности в бинарные метки
y_pred_bin = [1 if prob > 0.5 else 0 for prob in y_pred]
```

```python
# Вычисляем точность модели
accuracy = accuracy_score(y_test, y_pred_bin)
print('Точность модели:', accuracy)
```
